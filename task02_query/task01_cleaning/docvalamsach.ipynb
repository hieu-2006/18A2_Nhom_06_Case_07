```python
import pandas as pd
from pathlib import Path

# ==============================
# 1. ĐƯỜNG DẪN
# ==============================
# Thư mục chứa notebook: task01_cleaning
BASE_DIR = Path.cwd()

# Thư mục chứa file csv: ../data_raw
SOURCE_DIR = BASE_DIR.parent / "data_raw"

# Xuất file sạch ra data_clean
OUTPUT_DIR = BASE_DIR.parent / "data_clean"

# ==============================
# 2. ĐỌC DỮ LIỆU
# ==============================
hotel = pd.read_csv(SOURCE_DIR / "hotel_info.csv")
booking = pd.read_csv(SOURCE_DIR / "booking_records.csv")
review = pd.read_csv(SOURCE_DIR / "guest_review.csv")

# ==============================
# 3. LÀM SẠCH HOTEL_INFO
# ==============================
hotel["hotel_id"] = hotel["hotel_id"].astype(str).str.strip().str.upper() #Làm sạch hotel_id
hotel["hotel_name"] = hotel["hotel_name"].astype(str).str.strip().str.title() #Làm sạch hotel_name
hotel["city"] = hotel["city"].astype(str).str.strip().str.title() #Chuẩn hóa tên thành phố (city)
hotel["room_type"] = hotel["room_type"].astype(str).str.strip().str.title()#làm sạch room_
#Làm sạch base_price (loại ký tự tiền tệ, xử lý text)
hotel["base_price"] = (
    hotel["base_price"]
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)
)
hotel["base_price"] = pd.to_numeric(hotel["base_price"], errors="coerce")
hotel["base_price"] = hotel["base_price"].fillna(hotel["base_price"].mean())

# ==============================
# 4. LÀM SẠCH BOOKING_RECORDS
# ==============================
booking["hotel_id"] = booking["hotel_id"].astype(str).str.strip().str.upper()
booking["checkin_date"] = pd.to_datetime(booking["checkin_date"], errors="coerce")#Chuẩn hóa checkin_date về 1 định dạng
#Xử lý nights và num_guests sai định dạng hoặc âm
booking["nights"] = pd.to_numeric(booking["nights"], errors="coerce")
booking.loc[booking["nights"] <= 0, "nights"] = None
booking["nights"] = booking["nights"].fillna(1).astype(int)

booking["num_guests"] = pd.to_numeric(
    booking["num_guests"].astype(str).str.strip(),
    errors="coerce"
)
booking["num_guests"] = booking["num_guests"].fillna(1).astype(int)

# ==============================
# 5. LÀM SẠCH GUEST_REVIEW
# ==============================
#Xử lý rating sai kiểu hoặc thiếu
review["rating"] = pd.to_numeric(review["rating"], errors="coerce")
review.loc[(review["rating"] < 1) | (review["rating"] > 5), "rating"] = None
review["rating"] = review["rating"].fillna(review["rating"].mean())

review["comment"] = review["comment"].fillna("").astype(str).str.strip()

# 6. XUẤT FILE SẠCH (HIỆN TRONG VS CODE)
# ==============================
hotel.to_csv(OUTPUT_DIR / "hotel_info_clean.csv", index=False)
booking.to_csv(OUTPUT_DIR / "booking_records_clean.csv", index=False)
review.to_csv(OUTPUT_DIR / "guest_review_clean.csv", index=False)

print("✅ Đã xuất 3 file sạch ra thư mục project (VS Code Explorer)")


```
