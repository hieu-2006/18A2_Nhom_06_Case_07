4. Thống kê số khách trung bình mỗi phòng theo loại room_type

5. Tìm các khách sạn có base_price bất thường (quá cao hoặc quá thấp)

```python
q1 = hotel["base_price"].quantile(0.25)
q3 = hotel["base_price"].quantile(0.75)
iqr = q3 - q1

outliers = hotel[
    (hotel["base_price"] < q1 - 1.5 * iqr) |
    (hotel["base_price"] > q3 + 1.5 * iqr)
]
outliers
```
